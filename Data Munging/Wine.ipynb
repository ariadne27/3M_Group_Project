```python
import os
import csv
import re
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
file_to_load = "Resources/wine_data_convert.csv"

# Read Purchasing File and store into Pandas data frame
wine_data = pd.read_csv(file_to_load)
wine_data.sort_values('key')
wine_data.head()
```

```python
wine_data['wine'] = wine_data['winery']+' '+ wine_data['variety']
wine_data.drop(['designation','province', 'country', 'region_1', 'region_2'], axis = 1, inplace = True)
wine_data.head()
```

```python
wine_data['dash1'] = wine_data['description'].str.replace('-', ' ')
wine_data['dash2'] = wine_data['dash1'].str.replace('.', ' ')
wine_data['dash3'] = wine_data['dash2'].str.replace(',', ' ')
wine_data['dash4'] = wine_data['dash3'].str.replace(':', ' ')
wine_data['dash5'] = wine_data['dash4'].str.replace(';', ' ')
wine_data['dash'] = wine_data['dash5'].str.replace('!', ' ')
wine_data['split']= wine_data['dash'].str.split(' ')
wine_data.head()
```

```python
print(wine_data.iloc[0,8])
```

```python
# File to Load (Remember to Change These)
file_to_load2 = "Resources/aroma_list.csv"

# Read Purchasing File and store into Pandas data frame
aromas = pd.read_csv(file_to_load2)
aromas
```

```python
aroma_list = []
for stink in aromas['tier 3']:
    smellstr = str(stink)
    smell = smellstr.lower()
    aroma_list.append(smell)
aroma_list
```

```python
wine_list = []
attribute_list = []
score_list = []
price_list = []
for i in range(len(wine_data)): 
    word_list = wine_data.iloc[i,8]
#     for flavors in wine_data.iloc[i,8]:
    for word in word_list:
        unique = []
        if word in aroma_list:
            if word not in unique:
                unique.append(word)
                wordstr = str(word)
                wordlw = wordstr.lower()
                attribute_list.append(wordlw)
                wine_list.append(wine_data.iloc[i,6])
                score_list.append(wine_data.iloc[i,2])
                price_list.append(wine_data.iloc[i,3])
print("Done!")
```

```python
attribute_list
```

```python
wine_word = pd.DataFrame({
    "wine": wine_list,
    "attribute": attribute_list,
    "score": score_list,
    "price": price_list
})
wine_word
```

```python
wine_word.to_csv('Output/wine_words.csv')
```

```python
wine_counts = wine_word.groupby(['attribute']).count()
wine_counts.drop(['score', 'price'], axis = 1, inplace = True)
wine_counts
```

```python
wine_avg = wine_word.groupby('attribute').mean()
wine_avg.sort_values('score')
```

```python
wine_counts.to_csv('Output/wine_count.csv')
wine_avg.to_csv('Output/wine_avg.csv')
```
